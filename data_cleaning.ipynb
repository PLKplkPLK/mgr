```python
import numpy as np
import pandas as pd
```

```python
expected_results_file = '../y.csv'
```

Import

```python
y = pd.read_csv(expected_results_file, encoding='windows-1252', encoding_errors='replace', sep=';')
y = y.drop(columns=['lp', 'Type', 'Site', 'ID', 'Site_code', 'Season', 'Zone', 'Date', 'Time', 'Temperature', 'Snow', 'Behavior', 'Mature_ind', 'Young_ind', 'Sex', 'Comment'])
```

```python
species_cols = list(y.columns)
species_cols.remove('File')
species_cols.remove('RelativePath')
```

Polecat is broken

```python
y['polecat'].replace({'0': 0, '1': 1, '#NAZWA?': 0}, inplace=True)
y['polecat'].unique()
```

Delete rows, where there are couple of different species (including human) present

```python
y_invalid = y[y[species_cols].sum(axis=1) != y[species_cols].max(axis=1)]
y_invalid
```

```python
y = y[y[species_cols].sum(axis=1) == y[species_cols].max(axis=1)]
y
```

Paths names change

```python
y['RelativePath'].unique()
```

```python
y['RelativePath'] = y['RelativePath'].str.replace(r'?“', 'Ó', regex=False)
y['RelativePath'] = y['RelativePath'].str.replace(r'Pó?na', 'Późna', regex=False)
y['RelativePath'] = y['RelativePath'].str.replace(r'jesie?', 'jesień', regex=False)
y['RelativePath'] = y['RelativePath'].str.replace(r'?ƒ', 'Ń', regex=False)
y['RelativePath'] = y['RelativePath'].str.replace(r'?�', 'Ł', regex=False)
```

```python
y = y.reset_index(drop=True)
y.head(5)
```

Hot encoding to species column

```python
species_cols = list(y.columns)
species_cols.remove('File')
species_cols.remove('RelativePath')
y['species'] = y[species_cols].idxmax(axis=1)

# Optionally drop old one-hot columns
y = y.drop(columns=species_cols)
y
```

```python
y.species.unique()
```

```python
y.species.value_counts()
```

```python
y.to_csv('y_clean.csv')
```
