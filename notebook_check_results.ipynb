```python
import numpy as np
import pandas as pd
```

```python
#file_name = 'results_deepfaune-yolov8s_960_vit_large_patch14_dinov2.lvd142m.csv'
file_name = 'results_speciesnet_pl.csv'
file_path = 'results/' + file_name
```

```python
correct = pd.read_csv('results/correct.csv', index_col=0)
data = pd.read_csv(file_path)[['image', 'detected_animal']]
merged = pd.merge(correct, data, on='image', how='inner')
merged['correct'] = merged['present_animal'] == merged['detected_animal']
merged
```

```python
np.sum(merged['correct']) / len(merged)
```

```python
merged.to_csv('results/final/'+file_name)
```
