#### Counting

```python
import os
import pandas as pd
```

```python
animals = [
    'bird', 'red deer', 'hare', 'lynx', 'wolf', 'dog',
    'roe deer', 'wild boar', 'squirrel', 'marten', 'beaver',
    'fox', 'otter', 'badger', 'wildcat', 'polecat', 'bear', 'raccoon',
    'raccoon dog', 'weasel', 'bison', 'cat', 'moose',
    'fallow deer', 'stoat', 'mink'
]
animals.sort()
animals
```

```python
df = pd.DataFrame({'animal': [], 'file': []})

for animal in animals:
    folder_path = f'{animal}/pictures_cropped'
    files = os.listdir(folder_path)

    df = pd.concat([df, pd.DataFrame({'animal': animal, 'file': files})], ignore_index=True)
```
